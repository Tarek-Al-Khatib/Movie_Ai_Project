```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# Set up the WebDriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

url = 'https://www.pathe.fr/films'  # Replace with actual URL if different
driver.get(url)

time.sleep(5)  # Adjust if needed based on page load time

movie_elements = driver.find_elements(By.CLASS_NAME, 'poster')
movie_urls = []

for movie_element in movie_elements:
    try:
        # Retrieve the URL for each movie
        movie_link = movie_element.find_element(By.TAG_NAME, 'a')
        movie_url = movie_link.get_attribute('href')
        movie_urls.append(movie_url)
    except Exception as e:
        print(f"Error retrieving movie URL: {e}")

for url in movie_urls:
    print(f"Movie URL: {url}")
```

```python
# Initialize an empty list to store movie details
movies = []
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
# Loop over each movie URL
for movie_url in movie_urls:
    driver.get(movie_url)
    time.sleep(1) 
    try:

        movie_title = driver.find_element(By.TAG_NAME, 'h1').text

        subtitle_div = driver.find_element(By.CLASS_NAME, 'hero-film__subtitle')
        subtitles = subtitle_div.find_elements(By.TAG_NAME, 'p')
        duration = subtitles[0].text if len(subtitles) > 0 else 'N/A'
        genre = subtitles[1].text if len(subtitles) > 1 else 'N/A'

        picture_div = driver.find_element(By.CLASS_NAME, 'modal__cover')
        picture_url = picture_div.find_element(By.TAG_NAME, 'img').get_attribute('src')
        
        body_div = driver.find_element(By.CLASS_NAME, 'hero-film__body')


        cast_container = body_div.find_element(By.TAG_NAME, 'div').find_element(By.TAG_NAME, 'div')

        cast_paragraph = cast_container.find_element(By.TAG_NAME, 'p')
        cast_spans = cast_paragraph.find_elements(By.TAG_NAME, 'span')

        
                
        description = driver.find_element(By.CSS_SELECTOR, "p.hero-film__desc.ft-default.c-white.desktop-only").text
        
        
        cast = ', '.join([span.text for span in cast_spans])

        production_year = body_div.find_element(By.TAG_NAME, 'p').text

        movies.append({
            'title': movie_title,
            'duration': duration,
            'genre': genre,
            'picture': picture_url,
            'description': description,
            'cast': cast,
            'production_year': production_year
        })

    except Exception as e:
        print(f"Error retrieving details for {movie_url}: {e}")

    driver.back()
    time.sleep(1)  # Allow main page to load

for movie in movies:
    print(movie)

driver.quit()
```

```python
for movie in movies:
    print(f"UPDATE movies SET description = '{movie['description']}' WHERE title = '{movie['title']}';")
```
